```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DATA_DIR = '/home/iiticos/Desktop/Personal Docs/QC/IMC-Prosperity_2024/Data/round3/'

list = ['prices_round_3_day_2.csv','prices_round_3_day_1.csv','prices_round_3_day_0.csv']

df1 = pd.read_csv(DATA_DIR + list[0], sep=';')
df2 = pd.read_csv(DATA_DIR + list[1], sep=';')
df3 = pd.read_csv(DATA_DIR + list[2], sep=';')
```

```python
df = pd.concat([df1, df3, df2], ignore_index=True)
df
```

```python
# df = df1
```

```python
for i in ['ROSES','CHOCOLATE','STRAWBERRIES','GIFT_BASKET']:
    df_roses = df[df['product']==i]
    df_roses['mid_price'].plot()
    plt.title(i + ' Price')
    print("STD DEV : ", df_roses['mid_price'].std(), "MEAN : ", df_roses['mid_price'].mean())
    plt.show()
```
